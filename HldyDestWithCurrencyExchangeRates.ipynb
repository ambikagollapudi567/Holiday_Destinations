```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen,Request
```

```python
url = "http://www.thrillophilia.com/blog/best-places-to-visit-around-the-world/"
```

```python
request = Request(url)
response = urlopen(request)
html = response.read()
html_soup = BeautifulSoup(html , 'html.parser')
   
```

```python
print(html_soup)
```

```python
litag=html_soup.find_all(class_="postcontent content")
place=[]
Country=[]
for lines in range(0,len(litag)):
    litag1=litag[lines].find_all('li')
for lines in range(0,len(litag1)):
    split1=litag1[lines].text.split(',')
    if(len(split1)==2):
        place.append(split1[0])
        Country.append(split1[1])
    else:
        s1=split1[0]
        s2=s1.split(" ")
        
        l=len(s2)
        
        place.append(split1[0])
        Country.append(s2[l-1])

HldyDest=pd.DataFrame({'Place':place,'Country':Country})
print(HldyDest.head(5))
        
        


```

```python
HldyDest.to_csv("HolidayDestinations.csv")
```

```python
HD["CurrencyName"]="NA"
HD["CurrencyCode"]="NA"
HD["1.00 INR"] ="NA"
HD["inv 1 INR"] ="NA"
```

```python
CER=pd.read_csv('newCurrencyExchangeList.csv')
```

```python
print(CER.head(5))
```

```python
l1=len(CER)
l2=len(HD)
```

```python
CER['Country']=CER['Country'].astype(str)
for i in range(0,l2):
    for j in range(0,l1):
        a=HD['Country'][i]
        b=CER['Country'][j]
        a=a.strip()
        b=b.strip()
        c=a[0:5]
        d=b[0:5]
        e=a[-5:-1]
        f=b[-5:-1]
        if("India" in a or a=="Leh"):
            HD.loc[i,'CurrencyCode']="INR"
            HD.loc[i,'CurrencyName']="RUPEE"
            HD.loc[i,'1.00 INR']="1"
            HD.loc[i,'inv 1 INR']="1"
            break
        if(a==b or c==d  or e==f ):
            HD.loc[i,'CurrencyCode']=CER['CurrencyCode'][j]
            HD.loc[i,'CurrencyName']=CER['CurrencyName'][j]
            HD.loc[i,'1.00 INR']=CER['1.00 INR'][j]
            HD.loc[i,'inv 1 INR']=CER['inv 1 INR'][j]
            break
       
            
    
```
